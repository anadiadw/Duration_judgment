```python
## importing all libraries and dependencies
import pandas as pd
import numpy as np
import os
from os import listdir
import glob
from os.path import isfile, join
import shutil
import PIL
from PIL import Image
import random
from random import sample
import math
from heapq import merge
from random import *
import secrets
from scipy.stats import spearmanr
from scipy import stats
import matplotlib.pyplot as plt
```

```python
summary_dir = '/Users/ayesh/Desktop/Gradstuff/duration_judgement/summary/mturk_summary/emoscene_summaries/'
summary_files = [f for f in listdir(summary_dir)]
summary_files
```

```python
pt_ls = []
neumean = []
emomean = []
contmean = []
bdrymean = []

Ncont_mean = []
Econt_mean = []
Nbdry_mean = []
Ebdry_mean = []

for a in summary_files:
    pt_ls.append(a[:-4])
    pt_csv = pd.read_csv(summary_dir + a)
    pt_df = pt_csv[pt_csv["catch_trial"] == "No"]
    
    neupt = pt_df[pt_df["valence"] == "neutral"]
    emopt = pt_df[pt_df["valence"] == "negative"]
    conpt = pt_df[pt_df["event"] == "cont"]
    bdrpt = pt_df[pt_df["event"] == "bdry"]
    
    Cneupt = pt_df[pt_df["condition"] == "continious"]
    Cemopt = pt_df[pt_df["condition"] == "Negcont"]
    Bneupt = pt_df[pt_df["condition"] == "boundary"]
    Bemopt = pt_df[pt_df["condition"] == "Negbdr"]
    
    neu_trls = []
    emo_trls = []
    con_trls = []
    bdr_trls = []
    
    CN_trls = []
    CE_trls = []
    BN_trls = []
    BE_trls = []
    
    for index, row in neupt.iterrows():
        if str(neupt.slider_response[index]) != 'nan':
            neu_trls.append(abs(neupt.dur[index] - neupt.slider_response[index]))
    for index, row in emopt.iterrows():
        if str(emopt.slider_response[index]) != 'nan':
            emo_trls.append(abs(emopt.dur[index] - emopt.slider_response[index]))
    for index, row in conpt.iterrows():
        if str(conpt.slider_response[index]) != 'nan':
            con_trls.append(abs(conpt.dur[index] - conpt.slider_response[index]))
    for index, row in bdrpt.iterrows():
        if str(bdrpt.slider_response[index]) != 'nan':
            bdr_trls.append(abs(bdrpt.dur[index] - bdrpt.slider_response[index]))
    
    for index, row in Cneupt.iterrows():
        if str(Cneupt.slider_response[index]) != 'nan':
            CN_trls.append(abs(Cneupt.dur[index] - Cneupt.slider_response[index]))
    for index, row in Cemopt.iterrows():
        if str(Cemopt.slider_response[index]) != 'nan':
            CE_trls.append(abs(Cemopt.dur[index] - Cemopt.slider_response[index]))
    for index, row in Bneupt.iterrows():
        if str(Bneupt.slider_response[index]) != 'nan':
            BN_trls.append(abs(Bneupt.dur[index] - Bneupt.slider_response[index]))
    for index, row in Bemopt.iterrows():
        if str(Bemopt.slider_response[index]) != 'nan':
            BE_trls.append(abs(Bemopt.dur[index] - Bemopt.slider_response[index]))
            
    neumean.append(np.mean(neu_trls))
    emomean.append(np.mean(emo_trls))
    contmean.append(np.mean(con_trls))
    bdrymean.append(np.mean(bdr_trls))

    Ncont_mean.append(np.mean(CN_trls))
    Econt_mean.append(np.mean(CE_trls))
    Nbdry_mean.append(np.mean(BN_trls))
    Ebdry_mean.append(np.mean(BE_trls))
```

```python
emo_bdr_df = pd.DataFrame({"participant" :pt_ls,
                           "neutrals" : neumean,
                           "emotionals" :emomean,
                           "continious" : contmean,
                           "boundary" : bdrymean,
                           "Neu_cont": Ncont_mean,
                           "Emo_cont" : Econt_mean,
                           "Neu_bdr" : Nbdry_mean,
                           "Emo_bdr" : Ebdry_mean})
```

```python
emo_bdr_df.to_csv('/Users/ayesh/Desktop/Gradstuff/duration_judgement/summary/mturk_summary/r_analysis/abs_emo_bdr.csv')
```

```python
grph_csv = pd.read_csv('/Users/ayesh/Desktop/Gradstuff/duration_judgement/summary/mturk_summary/r_analysis/abs_emo_bdr.csv')
#grph_csv = pd.read_csv('/Users/ayesh/Desktop/Gradstuff/duration_judgement/summary/mturk_summary/r_analysis/emo_bdr.csv')
```

```python
Mgrph_csv = pd.melt(grph_csv, id_vars=['participant'], value_vars=['Neu_cont', 'Emo_cont', 'Neu_bdr', 'Emo_bdr'], var_name='condition', value_name='Total_acc')
Mgrph_csv.loc[Mgrph_csv['condition'] == 'Neu_cont', 'Test'] = 'Continious'
Mgrph_csv.loc[Mgrph_csv['condition'] == 'Emo_cont', 'Test'] = 'Continious'
Mgrph_csv.loc[Mgrph_csv['condition'] == 'Neu_bdr', 'Test'] = 'Boundary'
Mgrph_csv.loc[Mgrph_csv['condition'] == 'Emo_bdr', 'Test'] = 'Boundary'

Mgrph_csv.loc[Mgrph_csv['condition'] == 'Neu_cont', 'valence'] = 'c-neutral'
Mgrph_csv.loc[Mgrph_csv['condition'] == 'Emo_cont', 'valence'] = 'emotional'
Mgrph_csv.loc[Mgrph_csv['condition'] == 'Neu_bdr', 'valence'] = 'c-neutral'
Mgrph_csv.loc[Mgrph_csv['condition'] == 'Emo_bdr', 'valence'] = 'emotional'

Continious = Mgrph_csv.loc[Mgrph_csv['Test'] == 'Continious']
Boundary = Mgrph_csv.loc[Mgrph_csv['Test'] == 'Boundary']
C_Neutrals = Continious.loc[Continious['valence'] == 'c-neutral']
C_Emotionals = Continious.loc[Continious['valence'] == 'emotional']
B_Neutrals = Boundary.loc[Boundary['valence'] == 'c-neutral']
B_Emotionals = Boundary.loc[Boundary['valence'] == 'emotional']
```

```python
Neutrals_CB_mean = [np.mean(C_Neutrals['Total_acc']),np.mean(B_Neutrals['Total_acc'])]
Emotionals_CB_mean = [np.mean(C_Emotionals['Total_acc']), np.mean(B_Emotionals['Total_acc'])]
Neutrals_CB_sems = [stats.sem(C_Neutrals['Total_acc']), stats.sem(B_Neutrals['Total_acc'])]
Emotionals_CB_sems = [stats.sem(C_Emotionals['Total_acc']), stats.sem(B_Emotionals['Total_acc'])]
```
