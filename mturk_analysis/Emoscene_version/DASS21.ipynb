```python
#4-have an overall score for the three categories DAS for each participant (calculated in excel) and label them as normal, moderate, severe etc.
dep_csv = pd.read_csv('/Users/ayesh/Desktop/Gradstuff/duration_judgement/summary/mturk_summary/emoscene_questionnaires/dep_df.csv')
anx_csv = pd.read_csv('/Users/ayesh/Desktop/Gradstuff/duration_judgement/summary/mturk_summary/emoscene_questionnaires/anx_df.csv')
str_csv = pd.read_csv('/Users/ayesh/Desktop/Gradstuff/duration_judgement/summary/mturk_summary/emoscene_questionnaires/str_df.csv')

## for dep
dep_csv.loc[dep_csv['Dsum'] <= 9, 'Dlabel'] = 'Normal'
dep_csv.loc[(dep_csv['Dsum'] >= 10) & (dep_csv['Dsum'] <= 13), 'Dlabel'] = 'Mild'
dep_csv.loc[(dep_csv['Dsum'] >= 14) & (dep_csv['Dsum'] <= 20), 'Dlabel'] = 'Moderate'
dep_csv.loc[(dep_csv['Dsum'] >= 21) & (dep_csv['Dsum'] <= 27), 'Dlabel'] = 'Severe'
dep_csv.loc[dep_csv['Dsum'] >= 28, 'Dlabel'] = 'Extreme'

anx_csv.loc[anx_csv['Asum'] <= 7, 'Alabel'] = 'Normal'
anx_csv.loc[(anx_csv['Asum'] >= 8) & (anx_csv['Asum'] <= 9), 'Alabel'] = 'Mild'
anx_csv.loc[(anx_csv['Asum'] >= 10) & (anx_csv['Asum'] <= 14), 'Alabel'] = 'Moderate'
anx_csv.loc[(anx_csv['Asum'] >= 15) & (anx_csv['Asum'] <= 19), 'Alabel'] = 'Severe'
anx_csv.loc[anx_csv['Asum'] >= 20, 'Alabel'] = 'Extreme'

str_csv.loc[str_csv['Ssum'] <= 14, 'Slabel'] = 'Normal'
str_csv.loc[(str_csv['Ssum'] >= 15) & (str_csv['Ssum'] <= 18), 'Slabel'] = 'Mild'
str_csv.loc[(str_csv['Ssum'] >= 19) & (str_csv['Ssum'] <= 25), 'Slabel'] = 'Moderate'
str_csv.loc[(str_csv['Ssum'] >= 26) & (str_csv['Ssum'] <= 33), 'Slabel'] = 'Severe'
str_csv.loc[str_csv['Ssum'] >= 34, 'Slabel'] = 'Extreme'

dep_csv.to_csv('/Users/ayesh/Desktop/Gradstuff/duration_judgement/summary/mturk_summary/emoscene_questionnaires/dep_df.csv')
anx_csv.to_csv('/Users/ayesh/Desktop/Gradstuff/duration_judgement/summary/mturk_summary/emoscene_questionnaires/anx_df.csv')
str_csv.to_csv('/Users/ayesh/Desktop/Gradstuff/duration_judgement/summary/mturk_summary/emoscene_questionnaires/str_df.csv')
```
